```python
import pandas as pd
import numpy as np
```

```python
#implementing a 3 layer neural network(2 hidden layers)
#model parameters
input_size=2 # no of features
layers=[4,3] # no of neurons in first and second layer
output_size=2
```

```python
def softmax(a):
    e_pa=np.exp(a)#e^a
    ans=e_pa/np.sum(e_pa,axis=1,keepdims=True)
    return ans
```

```python
class NeuralNetwork:
    def __init__(self,input_size,layers,output_size):
        np.random.seed(0)#everytime initialse weights with same values
        model={}
        #layer 1
        model['w1']=np.random.randn(input_size,layers[0])#input with layer 1
        model['b1']=np.zeros((1,layers[0]))# bias of layer 1
        #layer 2
        model['w2']=np.random.randn(layers[0],layers[1])#layer1 with layer 2
        model['b2']=np.zeros((1,layers[1]))# bias of layer 2
        #layer 3 or output layer
        model['w3']=np.random.randn(layers[1],output_size)# layer 2 with ouput
        model['b3']=np.zeros((1,output_size))
        self.model=model
    def forward(self,x):
        w1,w2,w3=self.model['w1'],self.model['w2'],self.model['w3']
        b1,b2,b3=self.model['b1'],self.model['b2'],self.model['b3']
        z1=np.dot(x,w1)+b1
        a1=np.tanh(z1)
        z2=np.dot(a1,w2)+b2
        a2=np.tanh(z2)
        z3=np.dot(a2,w3)+b3
        y_=softmax(z3)#mx2
        self.activation_outputs=(a1,a2,y_)
        return y_
    def backward(self,x,y,learning_rate=0.001):
        w1,w2,w3=self.model['w1'],self.model['w2'],self.model['w3']
        b1,b2,b3=self.model['b1'],self.model['b2'],self.model['b3']
        m=x.shape[0]
        a1,a2,y_=self.activation_outputs
        delta3=y_- y
        #dw is derivative of loss function
        dw3=np.dot(a2.T,delta3)
        db3=np.sum(delta3,axis=0)
        delta2=(1-np.square(a2))*np.dot(delta3,w3.T)#derivative of sigmoid(z) which is tanh in this cases so derivative of tanh
        dw2=np.dot(a1.T,delta2)
        db2=np.sum(delta2,axis=0)
        
        delta1=(1-np.square(a1))*np.dot(delta2,w2.T)
        dw1=np.dot(x.T,delta1)
        db1=np.sum(delta1,axis=0)
        
        #update model parameters using gradient desecent
        self.model['w1'] -= learning_rate*dw1
        self.model['b1'] -= learning_rate*db1
        
        self.model['w2'] -= learning_rate*dw2
        self.model['b2'] -= learning_rate*db2
        
        self.model['w3'] -= learning_rate*dw3
        self.model['b3'] -= learning_rate*db3
    def predict(self,x):
        y_out=self.forward(x)
        return np.argmax(y_out,axis=1)
    def summary(self):
        w1,w2,w3=self.model['w1'],self.model['w2'],self.model['w3']
        b1,b2,b3=self.model['b1'],self.model['b2'],self.model['b3']
        a1,a2,y_=self.activation_outputs
        print("W1 ",w1.shape)
        print("A1 ",a1.shape)
        
        print("W2 ",w2.shape)
        print("A2 ",a2.shape)
        
        print("W3 ",w3.shape)
        print("y_ ",y_.shape)
```

```python
#cross entropy
def loss(y_oht,p):
    l = -np.mean(y_oht*np.log(p))
    return l
def one_hot(y,depth):
    m=y.shape[0]
    y_oht=np.zeros((m,depth))
    y_oht[np.arange(m),y]=1
    return y_oht
```

```python
from sklearn.datasets import make_circles,make_moons,make_classification
import matplotlib.pyplot as plt
```

```python
x,y=make_circles(n_samples=500,shuffle=True,noise=0.2,factor=0.2)
```

```python
plt.style.use('seaborn')
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Accent)
plt.show()
```

```python
print(y.shape)
y_oht=one_hot(y,2)
print(y_oht.shape)
```

```python
#Training our model
model=NeuralNetwork(input_size=2,layers=[10,5],output_size=2)
model.forward(x)
model.summary()
```

```python
def train(x,y,model,epochs,learning_rate,logs=True):
    training_loss=[]
    classes=10
    y_oht=one_hot(y,classes)
    for ix in range(epochs):
        y_=model.forward(x)
        l=loss(y_oht,y_)
        training_loss.append(l)
        model.backward(x,y_oht,learning_rate)
        if(logs):
            print(l)
    return training_loss        
```

```python
losses=train(x,y,model,500,0.001,True)
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
##Find accuracy
outputs=model.predict(x)
print(np.sum(outputs==y)/y.shape[0]*100,"%")
```

```python
#Effect of hidden units on different datasets
def load_dataset(dataset):
    if dataset=='moons':
        x,y=make_moons(n_samples=500,noise=0.2,random_state=1)
    elif dataset=='circles':
        x,y=make_circles(n_samples=500,shuffle=True,noise=0.2,random_state=1,factor=0.2)
    elif dataset=='classification':
        x,y=make_classification(n_samples=500,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=1)
    else:
        x=np.array([[0,0],
            [0,1],
            [1,0],
            [1,1]
           ])
        y=np.array([0,1,1,0])
    return x,y        
```

```python
datasets=['xor','classification','moons','circles']
xor_max=0
classs_max=0
moons_max=0
circles_max=0
xor_l1,xor_l2,classs_l1,classs_l2,moons_l1,moons_l2,circles_l1,circles_l2=0,0,0,0,0,0,0,0
for l1 in range(2,7):
    for l2 in range(2,7):
        for d in datasets:
            model=NeuralNetwork(input_size=2,layers=[l1,l2],output_size=2)
            x,y=load_dataset(d)
            train(x,y,model,1000,0.001,logs=False)
            outputs=model.predict(x)
            accuracy=np.sum(outputs==y)/y.shape[0]
            if(d=='xor'):
                if accuracy>xor_max:
                    xor_max=accuracy
                    xor_l1=l1
                    xor_l2=l2
            elif(d=='classification'):
                if accuracy>classs_max:
                    classs_max=accuracy
                    classs_l1=l1
                    classs_l2=l2
            elif(d=='moons'):
                if accuracy>moons_max:
                    moons_max=accuracy
                    moons_l1=l1
                    moons_l2=l2
            else:
                if accuracy>circles_max:
                    circles_max=accuracy
                    circles_l1=l1
                    circles_l2=l2
```

```python
print(xor_max,xor_l1,xor_l2)
print(classs_max,classs_l1,classs_l2)
print(moons_max,moons_l1,moons_l2)
print(circles_max,circles_l1,circles_l2)
```
